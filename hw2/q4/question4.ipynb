```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import eigs
import networkx as nx
```

```python
G = nx.random_graphs.extended_barabasi_albert_graph(50, 1, 0.2, 0.1,seed=10)
```

```python
K1 = nx.algorithms.approximation.steiner_tree(G,[10,14,3])
nx.draw(K1)
plt.savefig("q4/4.2.1.png")
nx.write_adjlist(K1,"q4/K1_adjacency_list.npy")
```
